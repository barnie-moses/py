```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

plt.scatter(x,y, color='red', marker="*", linewidths=5)
```

```python
def gradient_descent(x,y):
    m_slope = b_intercept = 0
    rate = 0.01
    n = len(x)
    plt.scatter(x,y, color='red', marker="*", linewidths=5)
    for i in range(10000):
        y_predicted = m_slope * x + b_intercept
        plt.plot(x,y_predicted, color='yellow')
        derivate_m = -(2/n) * sum(x*(y-y_predicted))
        derivate_b = -(2/n) * sum(y-y_predicted)
        m_slope = m_slope - rate * derivate_m
        b_intercept = b_intercept - rate * derivate_b
```

```python
gradient_descent(x,y)
```
